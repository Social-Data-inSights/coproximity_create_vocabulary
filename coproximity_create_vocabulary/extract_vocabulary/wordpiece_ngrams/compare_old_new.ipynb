```python
import json , os
```

```python
whole_folder :str = '../../../../data/wikipedia/whole/' 
save_folder = whole_folder+'wordpiece_based/'

old_file = save_folder + 'list_ngrams_ade_max_nb_1000000_min_likelihood_0.1.json'
new_file = save_folder + 'list_ngrams_ade_max_nb_1000000_min_likelihood_0.1_test10.json'
```

```python
with open(old_file) as f :
    old = json.load(f)
    
with open(new_file) as f :
    new_ = json.load(f)
```

```python
dict_old = {(first , second) : (freq, likelihood) for first , second , freq, likelihood in old}
dict_new = {(first , second) : (freq, likelihood) for first , second , freq, likelihood in new_}
```

```python
nb = 0
for first , second , freq, likelihood in old :
    if not (first,second) in dict_new :
        nb += 1
        print(first,second , (freq, likelihood)  , 'missing' )
nb, len(old)
```

```python
nb = 0
for first , second , freq, likelihood in new_ :
    if not (first,second) in dict_old :
        nb += 1
        print(first,second , (freq, likelihood)  , 'missing' )
nb, len(new_)
```
